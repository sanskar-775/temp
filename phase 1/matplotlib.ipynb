```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Line plot

```python
x = [2,5,6,7,8]
y = [3,7,4,5,12]
plt.plot(x,y)
plt.title("line graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y,'^')
plt.title("line graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y,'^')
plt.title("line graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.arange(0,5*np.pi,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y,)
plt.title("sine graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 2. Scatter plot

```python
x = [2,5,6,7,8]
y = [3,7,4,5,12]

plt.scatter(x,y)

plt.title("Scatter graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 3. Comparison

```python
x = [2,5,6,7,8]
y = [3,7,4,5,12]

x2 = [4,6,9,2,4]
y2 = [4,5,2,9,6]

plt.plot(x,y,color='g',linewidth=8)
plt.plot(x2,y2)

plt.legend(["First_line","Second_Line"],loc=4)

plt.title("Comparison graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 4. Bar Graph

```python
x = [2,5,6,7,8]
y = [3,7,4,5,12]

x2 = [4,6,9,2,4]
y2 = [4,5,2,9,6]

plt.bar(x,y,color='g',linewidth=8)
plt.bar(x2,y2)

plt.legend(["First_line","Second_Line"],loc=4)

plt.title("Comparison graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [2,5,6,7,8]
y = [3,7,4,5,12]

x2 = [4,6,9,2,4]
y2 = [4,5,2,9,6]

plt.barh(x,y,color='g',linewidth=8)
plt.barh(x2,y2)

plt.legend(["First_line","Second_Line"],loc=4)

plt.title("Comparison graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 5. Bubble Graph

```python
x = [2,5,6,7,8]
y = [3,7,4,5,12]

plt.scatter(x,y, s = np.random.rand(5)*1000, c=np.random.rand(5))

plt.title("Scatter graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 6. Pie Chart

```python
data = [3,7,4,12,5]
labels = ['a','b','c','d','e']
ex = [0,0.1,0,0,0]
plt.pie(data,labels=labels,explode=ex,shadow=True)
plt.show()
```
